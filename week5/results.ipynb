## Discussion
### Gradient Descent
Really only worth commenting on the case where learning rate = 0.001.
Here, the algorithm does not seem to move large enough steps in the gradient direction, and therefore does not learn quickly enough. It both takes a while (1000+ iterations) to start seeing improvement, and after it flattens out the improvement is - obviously - not as quick as the others to reach near-optimal, and that's why it ends at 7.56% misclassification rate.

It's also worth noting that Gradient Descent was the algorithm that generally took the longest, out of the gradient methods we tried, whether that was with momentum/weight decay or without, and that for larger step sizes, it took shorter time.

#### With Momentum
Momentum seems also generally to take its time, although we see an abrupt - nearly qualitative change between momentum learning rate 0.001 when alpha $= 0.8$ compared to $0.9$. I can't really explain what's happening there, other than that we started overfitting earlier than in the $0.8$ case.

However, that seems also to generally be the case with momentum: it aborts early (due to us including early stopping with our momentum, which might have been a mistake, but this kind of experimentation is how you learn, isn't it?).

#### With Weight Decay
Weight decay received similar results as momentum and general GD.
